```python
import pandas as pd
import numpy as np

path = "toorandmed.csv"

df = pd.read_csv(path)

df.head()

n_andmepunktid = df.shape[0]   # rows
n_tunnused = df.shape[1]       # columns

print("Andmepunktide arv (ridu):", n_andmepunktid)
print("Tunnuste arv (veerge):", n_tunnused)

print("\nTunnuste nimed:")
print(list(df.columns))


# pick a random row
random_index = np.random.randint(0, len(df))
row = df.iloc[random_index]

print(f"Valitud juhuslik rida (index = {random_index})")

# print Estonian title
print("\nAine nimetus (eesti keeles):")
print(row["title__et"])

empty_count = row.isna().sum()
total_fields = len(row)

print("\nPuuduvate väärtuste arv selles reas:")
print(f"{empty_count} / {total_fields}")
print(f"Puuduvate väärtuste osakaal: {empty_count / total_fields:.2%}")

dtype_summary = {}

for col in df.columns:
    if not pd.isna(row[col]):
        dtype = str(df[col].dtype)
        dtype_summary[dtype] = dtype_summary.get(dtype, 0) + 1

dtype_summary
```
