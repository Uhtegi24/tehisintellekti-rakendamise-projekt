## Vigade analüüsi kokkuvõte

Analüüsisime vigaseid päringuid ning tuvastasime, millises vahesammus viga tekkis.

Kokku vigaseid päringuid: **3**

| Vahesamm                           | Vigade arv | % kõikidest vigastest päringutest |
|------------------------------------|------------|-----------------------------------|
| Metaandmete filtreerimine         | 0          | 0%                                 |
| RAG vektorotsing                  | 2          | 50%                                |
| LLM vastuse genereerimine         | 2          | 50%                                |
| **Kokku**                         | **4**     | **100%**                            |
